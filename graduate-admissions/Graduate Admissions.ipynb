```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
Graduate = pd.read_csv('Admission_Predict.csv')
# I import the csv file into to pandas to convert the file into a dataframe that can be worked with.
```

```python
Graduate = Graduate.drop('Serial No.', axis = 1)
Graduate
#I see that Serial No. represents the students. Since each student is different, the column can be used as the index because it 
# uniquely identifies the 
```

```python
Graduate1 = Graduate[Graduate.CGPA >= 0.85]
Graduate1.plot(kind = 'scatter', x = 'GRE', y='TOEFL')
plt.title('GRE vs TOEFL scores with good GPA')
plt.grid(True)
```

```python
def binary(Graduate):
    if Graduate['Admit']>0.7:
        return 1
    else:
        return 0
Graduate['Admit'] = Graduate.apply(binary, axis=1)
Graduate1['Admit'] = Graduate1.apply(binary, axis=1)
```

```python
ax = sns.scatterplot(data = Graduate1, x = 'GRE', y= 'CGPA', hue = 'Admit')
plt.title('Scores for Admission(GRE VS CGPA)')
plt.grid(True)
```

```python
ax = sns.scatterplot(data = Graduate1, x = 'TOEFL', y= 'CGPA', hue = 'Admit')
plt.title('Scores for Admission(TOEFL VS CGPA)')
plt.grid(True)
```

```python
ax = sns.scatterplot(data = Graduate1, x = 'TOEFL', y= 'GRE', hue = 'Admit')
plt.title('Scores for Admission(GRE VS TOEFL)')
plt.grid(True)
```
